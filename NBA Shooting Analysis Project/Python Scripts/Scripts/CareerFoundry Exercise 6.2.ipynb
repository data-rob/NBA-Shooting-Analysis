## 3. Create a correlation heatmap using seaborn:

```python
#Viewing data columns

data.columns
```

```python
# Create a subset excluding the "Player", "Pos", "Season", "Tm", & "Player-additional" columns

sub = data[['Pts Rnk', 'Age', 'G', 'GS', 'MP',
       'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%',
       'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV',
       'PF', 'PTS']]
```

```python
sub
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(20,14))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax, cmap="PiYG") # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

The green colors in the heatmap indicate that there is a positive relationship between the two variables that were crossed. The darker the green color means the stronger the relationship is. Pink is used to signify a negative relationship between the two variables and the darker the pink color means the stronger the relationship.

## 7. Answering questions

```python
average_ft_per_by_position = data.groupby('Pos')['FT%'].mean().reset_index()
```

```python
average_ft_per_by_position = average_ft_per_by_position.sort_values('FT%', ascending=False)
```

```python
f,ax = plt.subplots(figsize=(20,14))
sns.barplot(data = average_ft_per_by_position, x = average_ft_per_by_position['Pos'], y = average_ft_per_by_position['FT%'], errorbar=None)
```

Here is the visualization to the question of how different positions compare in their free throw shooting percentage. The guard positions tend to have better percentages from the free throw line than the forwards. This could be used as evidence to support the claim that size affects shooting because the bigger players at the forward and center positions do not make as many of their free throws as the smaller players do. Free throw percentage is a very good indicator of how good a player is at shooting because the free throw shot is identical for every player in every situation in a physical sense.
